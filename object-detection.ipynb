```python
import cv2
```

```python
weights="C:\\Users\\kasuk\\OneDrive\\Desktop\\python files\\yolov3-tiny.weights"
cfg="C:\\Users\\kasuk\\OneDrive\\Desktop\\python files\\yolov3-tiny.cfg.txt"
classes="C:\\Users\\kasuk\\OneDrive\\Desktop\\python files\\coco_classes.txt"
```

```python
net=cv2.dnn.readNet(weights,cfg)
model=cv2.dnn_DetectionModel(net)
model.setInputParams(size=(155,155),scale=1/255)
```

```python
classes=[]
with open("C:\\Users\\kasuk\\OneDrive\\Desktop\\python files\\coco_classes.txt",'r') as file_object:
    for class_name in file_object.readlines():
        #print(class_name)
        class_name=class_name.strip()
        classes.append(class_name)
print(classes)
        
```

```python
cam=cv2.VideoCapture(0)
cam.set(cv2.CAP_PROP_FRAME_WIDTH,1280)
cam.set(cv2.CAP_PROP_FRAME_HEIGHT,720)
```

```python
while True:
    ret,frame=cam.read()
    # object detection
    class_ids,scores,bboxes=model.detect(frame)
    for class_id, score, bbox in zip(class_ids,scores,bboxes):
        (x,y,w,h)=bbox
        class_name=classes[class_id]
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
        cv2.putText(frame,class_name,(x,y-5),cv2.FONT_HERSHEY_PLAIN,1,(0,0,255),2)
        cv2.putText(frame,str(score),(x,y-25),cv2.FONT_HERSHEY_PLAIN,1,(0,0,255),2)
    print('class_id',class_ids)
    print('score',scores)
    print('bbox',bboxes)
    if cv2.waitKey(1)==ord('q'):
        break
    cv2.imshow('frame',frame)
    
cam.release()
```
